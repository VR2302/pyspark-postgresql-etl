```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PostgreSQL Connection") \
    .master("local[*]") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.7.3") \
    .config("spark.driver.host", "127.0.0.1") \
    .getOrCreate()

print("✅ Spark Working vipin")
```

```python
df = spark.read.csv("file:///D:\PYTHON\8_version\employees.csv", header=True, inferSchema=True)
df.show(5)
```

```python
df = spark.read.csv("file:///D:\PYTHON\8_version\employees.csv", header=True, inferSchema=True)
df.show(10)
```

```python
df_filtered = df.filter(df.salary > 50000)

df_filtered.show()
```

```python
df_grouped = df.groupBy("department").count()

df_grouped.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

windowSpec = Window.partitionBy("department").orderBy(df.salary.desc())

df_ranked = df.withColumn("rank", row_number().over(windowSpec))

df_ranked.show()
```

```python
url = "jdbc:postgresql://localhost:5432/Test_DB"

properties = {
    "user": "postgres",
    "password": "abc123",
    "driver": "org.postgresql.Driver"
}
```

```python
df_ranked.write \
    .jdbc(url=url, table="emp_ranked", mode="overwrite", properties=properties)
print("✅ Data successfully written emp_ranked to PostgreSQL!")
```

```python
df.write \
    .jdbc(url=url, table="employees", mode="overwrite", properties=properties)

print("✅ Data successfully written to PostgreSQL!")
```
